#**IRIS DATASETS**

**scatter plot**

```python
import pandas as pd
import matplotlib.pyplot as plt
file_name='/Iris.csv'
df=pd.read_csv('/content/Iris.csv')
color={'Iris-setosa':'red','Iris-virginica':'blue' , 'Iris-versicolor':'green'}
sp_color = df['Species'].map(color)
df.columns=['Id', 'sepal_Length','sepal_Width','petal_Length','petal_Width','class']
print(df.columns)
x = df['sepal_Length']
y = df['sepal_Width']
plt.scatter(x,y, c=sp_color)
plt.xlabel('sepal_Length in CM')
plt.ylabel('sepal_Width in CM')
plt.title('scater plot')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
file_name='/Iris.csv'
df=pd.read_csv('/content/Iris.csv')
df.columns=['Id', 'sepal_Length','sepal_Width','petal_Length','petal_Width','class']
print(df.columns)
plt.boxplot(df['sepal_Length'])
plt.xlabel('sepal_Length in CM')
plt.title('box plot')
```

**bar plot**

```python
import pandas as pd
import matplotlib.pyplot as plt
file_name='/Iris.csv'
df=pd.read_csv('/content/Iris.csv')
df.columns=['Id', 'sepal_Length','sepal_Width','petal_Length','petal_Width','class']
species_count=df['class'].value_counts()
plt.bar(species_count.index,species_count.values,color='skyblue')
plt.xlabel('species')
plt.ylabel('count')
plt.title('bar plot')
plt.show()
```

#**student information**


```python
import pandas as pd
import matplotlib.pyplot as plt
file_name='/StudentsPerformance.csv'
df=pd.read_csv('/content/StudentsPerformance.csv')
genders = df['gender'].unique()
data = [df[df['gender'] == g]['math score'].values for g in sorted(genders)]
gender_labels = sorted(genders)
plt.boxplot(data, labels=gender_labels)
plt.title('Box Plot of Math Score by Gender')
plt.xlabel("Gender")
plt.ylabel("Math Score")
plt.grid(True)
plt.show()
```
