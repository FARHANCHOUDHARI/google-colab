```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('/content/Book1.csv')
plt.hist(df['CIE-I'],edgecolor='black')
plt.xlabel('CIE-I Score')
plt.ylabel('Frequency')
plt.title('Distribution of CIE-I Scores')
plt.show()
```

```python
plt.scatter(df['CIE-I'],df['CIE-II'])
plt.xlabel('CIE-I Score')
plt.ylabel('CIE-II Score')
plt.title('CIE-I vs CIE-II Scores')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('/content/Book1.csv')
plt.boxplot([df['CIE-I'],df['CIE-II']])
plt.xticks([1,2],['CIE-I','CIE-III'])
plt.ylabel('Score')
plt.title('Boxplot of CIE-I and CIE-II Scores')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('/content/Netflix.csv')
print(df.describe)
plt.show()
```

```python
plt.hist(df['release_year'],edgecolor='black')
plt.xlabel('Release Year')
plt.ylabel('Frequency')
plt.title('Distribution of Release Years')
plt.grid(True)
plt.show()
```

```python
plt.scatter(df['release_year'],df['date_added'])
plt.xlabel('Release Year')
plt.ylabel('Date Added')
plt.title('Release Year vs Date Added')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('/content/Netflix.csv')
df['date_added'] = pd.to_datetime(df['date_added'])
plt.boxplot([df['release_year'], df['date_added'].dt.year])
plt.xticks([1,2],['Release Year','Year Added'])
plt.ylabel('Year')
plt.title('Boxplot of Release Year and Year Added')
plt.grid(True)
plt.show()
```
